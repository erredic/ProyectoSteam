```python
# Cargar los archivos CSV en pandas DataFrames
DFSteamGames = pd.read_csv('SteamGames.csv')
DFUserItems = pd.read_csv('UsersItems.csv')

# Filtrar los juegos por género y obtener el usuario con más horas jugadas
def obtener_usuario_mas_horas_jugadas(genero):
    # Eliminar filas con valores NaN en la columna 'genres'
    DFSteamGamescleaned = DFSteamGames.dropna(subset=['genres'])
    
    # Filtrar los juegos por el género dado
    juegos_genero = DFSteamGamescleaned[DFSteamGamescleaned['genres'].str.contains(genero, case=False)]
    
    # Verificar si hay juegos asociados al género dado
    if juegos_genero.empty:
        return f"No hay juegos asociados al género '{genero}'."
    
    # Unir los DataFrames para obtener las horas jugadas por usuario
    horas_jugadas_por_usuario = pd.merge(juegos_genero, DFUserItems, on='item_id')
    
    # Sumar las horas jugadas por usuario
    horas_por_usuario = horas_jugadas_por_usuario.groupby('user_id')['playtime_forever'].sum().reset_index()
    
    # Verificar si hay datos para el género dado
    if horas_por_usuario.empty:
        return f"No hay datos disponibles para el género '{genero}'."
    
    # Obtener el usuario con más horas jugadas
    usuario_mas_horas = horas_por_usuario.loc[horas_por_usuario['playtime_forever'].idxmax()]
    
    return usuario_mas_horas

# Calcular la acumulación de horas jugadas por año de lanzamiento
def acumulacion_horas_por_anio():
    # Unir los DataFrames para obtener el año de lanzamiento de cada juego
    juegos_con_anio = pd.merge(DFSteamGames, DFUserItems, on='item_id')
    
    # Calcular las horas jugadas por año de lanzamiento
    horas_por_anio = juegos_con_anio.groupby('release_date')['playtime_forever'].sum().reset_index()
    
    return horas_por_anio.to_dict('records')


# Solicitar al usuario que ingrese el género
genero_dado = input("Ingrese el género: ")

# Obtener el usuario con más horas jugadas para el género dado
usuario_mas_horas_jugadas = obtener_usuario_mas_horas_jugadas(genero_dado)

# Calcular la acumulación de horas jugadas por año de lanzamiento
acumulacion_horas_anio = acumulacion_horas_por_anio()

# Crear el diccionario de retorno
retorno = {
    "Usuario con más horas jugadas para " + genero_dado: usuario_mas_horas_jugadas['user_id'],
    "Horas jugadas": acumulacion_horas_anio
}

# Imprimir el resultado
print(retorno)
```
